```python
## create list with hourly dates in summer 2014 

def daterange(start_date, end_date):
    delta = timedelta(hours=1)
    while start_date < end_date:
        yield start_date
        start_date += delta
```

```python
from datetime import datetime, timedelta
date_list=[]

start_date = datetime(2014, 6, 1, 00, 00)
end_date = datetime(2014, 8, 31, 23, 30)
for single_date in daterange(start_date, end_date):
    date_list.append(single_date.strftime("%Y-%m-%d %H:%M"))
```

```python
# read in text files 
import pandas as pd
import re

def get_data(file):
    act = 0 # activating variable to read in lines to df
    key= 'precipitationCal'   # key word after which lines should be extracted 
    numbers=[]
    i=0 
    
    with open(file) as cc: 
        for line in cc:   # wrap pbar around object to display progress 
            i+=1
            if key in line: 
                act = 1
            if act == 1:
                rate=re.findall(r'\b\d+\b',str(line)) # only select integers from string 
                if len(rate) > 0:
                    if len(rate) > 1:
                        decimal= '0.'+ rate[1]
                        numbers.append(decimal)
                    else:
                        numbers.append(rate[0])    
            if ';' in line: 
                act = 0   
    return numbers,i
```
